# Find delta

```python
# 
lst = np.random.randint(0,100,10)
lst = [3,4,5]

def find_delta(lst):
    delta_lst = []
    for i in range(0,len(lst),1):
        delta_lst.append(lst[i]- lst[i+1] )    
    return delta_lst
```

```python
print(lst)
```

```python
print(find_delta(lst))
```
