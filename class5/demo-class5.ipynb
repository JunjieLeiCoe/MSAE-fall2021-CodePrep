> comprehension = [ @return **data** for iterator(i) in **data-type**]

> comprehension = [ @return **data** for iterator(i) in **data-type** if statement == True]

> comprehension = [ @return **data** if statement == True else return **data** for iterator(i) in **data-type**]

```python
lst1 = [random.randint(0,100) for i in range(0,10,1)]
print(lst1)
orders = [[o,i] for o, i in enumerate(lst1)]
orders
```

```python
lst1 = [random.randint(0,100) for i in range(0,10,1)]
lst2 = [random.randint(0,100) for i in range(0,10,1)]

combined_lst = [[i,j] for i in lst1 for j in lst2 if i % 2 == 0 and j % 2 == 0]
print(combined_lst)
```

## Dictionary Comprehension

```python
dial_codes = [ (86, 'China'), (91, 'India'), (1, 'United States'), (62, 'Indonesia'),
               (55, 'Brazil'), (92, 'Pakistan'), (880, 'Bangladesh'), (234, 'Nigeria'),
               (7, 'Russia'), (81, 'Japan')]

print(dial_codes, "\n**", isinstance(dial_codes, list), "**", type(dial_codes))
```

```python
country_code  = {country: code for code, country in dial_codes} 

print(country_code, "\n**", isinstance(country_code, list), "**", type(country_code))
```
